```python
class StudentCard():
    def __init__(self, roll_no: int, name: str, age: int) -> None:
        if age < 18 or age > 60:
            raise StudentAgeError("You are not eligible!")
        self.roll_no = roll_no
        self.name = name
        self.age = age

# This is our student Error Class derived from Parent class(EXception)
class StudentAgeError(Exception):
    pass

student1 = StudentCard(1,"Mohammad Abbas",19) # There is no error
print(student1.roll_no, student1.name, student1.age)
```

```python
with open("./file.txt", 'r') as file:  # Type: TextIO
    print(file.read())
    file.write("This will generate error") # Not access to write
```

```python
with open("./file.txt", 'w') as file:  # Type: TextIO
    print(file.read()) # Not access to read
    file.write("This will not generate error") 
```

```python
with open("./file2.txt", 'w') as file:  # Type: TextIO
    file.write("This is Modern Python & this mode will override the existing data in the file") 
```

```python
with open("./file2.txt", 'r+') as file:  # Type: TextIO
    print(file.read()) # r+ mode has both access i.e read & write
    file.write("|The default cursor after first line will be set at the end of line & this will not be printed only will be written in file") 
```

```python
with open("./file3.txt", 'r+') as file:  # Type: TextIO
    print(file.read()) 

    file.write("Write after it")

    print("After: ", file.read())
```

```python
with open("./file3.txt", 'r+') as file:  # Type: TextIO
    print(file.read()) 

    file.write("Write after it")
    file.seek(0) # we can set our cursor position at 0 to read complete file after writing with the help of seek() function

    print("After: ", file.read())
```

```python
# !pip install PyPDF2
# Install the req package 

from typing import List
import PyPDF2

def read_pdf(file_path: str) -> List[str]:
    with open(file_path, 'rb') as file:  # The 'b' in 'rb' stands for binary mode
        reader: PyPDF2.PdfReader = PyPDF2.PdfReader(file)
        text_content: List[str] = [page.extract_text() for page in reader.pages]
        return text_content

pages: list[str] = read_pdf("./mypdf.pdf")
pages
```
